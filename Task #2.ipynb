```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
iris
```

```python
iris_data = pd.read_csv(r'C:\Users\tarun\Desktop\Iris.csv')
iris_data
```

```python
label_mapping = {
    'Iris-setosa':0,
    'Iris-versicolor':1,
    'Iris-virginica':2
}
iris_data['Target']=0
iris_data['Target']=iris_data['Species'].map(label_mapping)
iris_data.rename(columns={"Species":"Target"})
iris_data
```

```python
plt.scatter(x=iris_data['SepalLengthCm'],y=iris_data['SepalWidthCm'] ,c=iris_data.Target)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
km = KMeans(n_clusters=4)
y_predicted = km.fit_predict(iris_data[['SepalLengthCm','SepalWidthCm']])
y_predicted
```

```python
iris_data['cluster']=y_predicted
iris_data.head()
```

```python
km.cluster_centers_
```

```python
df1=iris_data[iris_data.cluster==0]
df2=iris_data[iris_data.cluster==1]
df3=iris_data[iris_data.cluster==2]
df4=iris_data[iris_data.cluster==3]
plt.scatter(df1.SepalLengthCm,df1.SepalWidthCm,color='r')
plt.scatter(df2.SepalLengthCm,df2.SepalWidthCm,color='g')
plt.scatter(df3.SepalLengthCm,df3.SepalWidthCm,color='b')
plt.scatter(df4.SepalLengthCm,df4.SepalWidthCm,color='y')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker="D")
```

```python
sse = []
k_range=range(1,10)
for k in range(1,10):
    km = KMeans(n_clusters=k)
    km.fit(iris_data[['SepalLengthCm','SepalWidthCm']])
    sse.append(km.inertia_)
    
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_range,sse)
```
